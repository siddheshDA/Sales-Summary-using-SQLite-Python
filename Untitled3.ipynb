```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# 1. Connect to the database
# ===============================
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()

# ===============================
# 2. Create sales table if not exists
# ===============================
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product TEXT,
    quantity INTEGER,
    price REAL
)
""")

# ===============================
# 3. Insert sample data (only if table is empty)
# ===============================
cursor.execute("SELECT COUNT(*) FROM sales")
if cursor.fetchone()[0] == 0:
    sample_data = [
        ('Product A', 10, 50),
        ('Product B', 5, 100),
        ('Product C', 8, 75),
        ('Product A', 7, 50),
        ('Product B', 3, 100),
        ('Product C', 6, 75)
    ]
    cursor.executemany(
        "INSERT INTO sales (product, quantity, price) VALUES (?, ?, ?)",
        sample_data
    )
    conn.commit()

# ===============================
# 4. SQL query for summary
# ===============================
sql_query = """
SELECT product,
       SUM(quantity) AS total_qty,
       SUM(quantity * price) AS revenue
FROM sales
GROUP BY product
"""

print("\nðŸ“Œ SQL Query Used:\n", sql_query)

# Read query result into Pandas DataFrame
df = pd.read_sql_query(sql_query, conn)

# ===============================
# 5. Print results in a nice format
# ===============================
print("=" * 40)
print("        ðŸ“Š SALES SUMMARY ðŸ“Š")
print("=" * 40)
print(df.to_string(index=False))
print("=" * 40)

# ===============================
# 6. Plot attractive bar chart
# ===============================
plt.figure(figsize=(6,4))
bars = plt.bar(df['product'], df['revenue'], color=['#4CAF50', '#2196F3', '#FF9800'])
plt.title("Revenue per Product", fontsize=14, fontweight='bold')
plt.ylabel("Revenue", fontsize=12)
plt.xlabel("Product", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels above bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 5,
             f"{bar.get_height():.0f}",
             ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig("sales_chart.png")
plt.show()

# ===============================
# 7. Close the connection
# ===============================
conn.close()
```
